```python
# Base path to the SalesLT folder
base_path = "/mnt/gold/SalesLT/"

# Switch to the database where you want to create the tables
spark.sql("CREATE DATABASE IF NOT EXISTS gold_db")
spark.sql("USE gold_db")

# Iterate over each table directory, infer schema, and create table in gold_db
for table in table_name:
    table_path = base_path + table
    # Load data from Delta or CSV (or change to format that matches your source)
    df = spark.read.format("delta").load(table_path)
    
    # Display the inferred schema
    print(f"Schema for {table}:")
    df.printSchema()

    # Write the DataFrame to a new SQL table in gold_db
    df.write \
      .format("jdbc") \
      .option("url", "jdbc:sqlserver://ddutjnrevenge-server.database.windows.net:1433;database=gold-db") \
      .option("dbtable", table) \
      .option("user", "") \
      .option("password", "") \
      .option("driver", "com.microsoft.sqlserver.jdbc.SQLServerDriver") \
      .mode("overwrite") \
      .save()

    print(f"Table {table} created in gold-db.\n")
```

```python
# Iterate over each table directory, infer schema, and create view in gold_db
# Base path to the SalesLT folder
base_path = "/mnt/gold/SalesLT/"

# Switch to the database where you want to create the tables
spark.sql("CREATE DATABASE IF NOT EXISTS gold_db")
spark.sql("USE gold-db")

for table in table_name:
    table_path = base_path + table
    # Load data from Delta format
    df = spark.read.format("delta").load(table_path)
    
    # Display the inferred schema
    print(f"Schema for {table}:")
    df.printSchema()

    # Create or replace a view in Spark
    view_name = f"{table}_view"  # Create a view name that avoids conflicts
    df.createOrReplaceTempView(view_name)

    print(f"View {view_name} created.\n")

# Optional: Display all views created
print("List of views created:")
views = spark.catalog.listTables()
for view in views:
    print(view.name)
```
